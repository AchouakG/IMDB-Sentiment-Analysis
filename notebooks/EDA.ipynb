```python
import pandas as pd

df = pd.read_csv('../data/bronze/IMDB Dataset.csv')

df.head(10)
```

```python
text_col = "review"

df[text_col] = df[text_col].astype(str)
empty = df[text_col].str.strip().eq("") | df[text_col].isna()
dup = df.duplicated(subset=[text_col])

empty.sum(), dup.sum()
```

```python
y_col="sentiment"

df[y_col].value_counts(normalize=True) # 50% positive 50% negative --> highly balanced
```

```python
s = df[text_col].fillna("").astype(str)
df["n_words"] = s.str.split().str.len()
df["n_chars"] = s.str.len()

df[["n_words","n_chars"]].describe(percentiles=[.5,.9,.95,.99]).T
# ! what does this do exactly ???? 
```

```python
import re

def clean_for_sentiment(t: str) -> str:
    t = str(t).lower()

    # to remove urls
    t = re.sub(r"http\S+|www\.\S+", " ", t)

    # to remove html tags
    t = re.sub(r"<.*?>", " ", t)

    # to keep punctuation, just normalize repeated spaces
    t = re.sub(r"\s+", " ", t).strip()
    return t

df["review"] = df[text_col].apply(clean_for_sentiment)
df["review"] = df["review"].str.lower()
df[["review"]].head()
```
