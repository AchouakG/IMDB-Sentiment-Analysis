```python
import pandas as pd
from pathlib import Path

RAW_PATH = "../data/raw/IMDB Dataset.csv"
CLEANED_PATH = "../data/clean/cleaned_imdb.csv"

Path("../data/clean").mkdir(parents= True, exist_ok=True)


df = pd.read_csv(RAW_PATH)

df.head(10)
```

```python
# checking for empty strings and duplicates

text_col = "review"

df[text_col] = df[text_col].astype(str)
empty = df[text_col].str.strip().eq("") | df[text_col].isna()
dup = df.duplicated(subset=[text_col])

empty.sum(), dup.sum()
```

```python
y_col="sentiment"

df[y_col].value_counts(normalize=True) # 50% positive 50% negative --> highly balanced
```

```python

# cleaning from urls and html tags and normalize repeated spaces

import re

def clean_for_sentiment(t: str) -> str:
    t = str(t).lower()

    t = re.sub(r"http\S+|www\.\S+", " ", t)

    t = re.sub(r"<.*?>", " ", t)

    t = re.sub(r"\s+", " ", t).strip()
    return t

df["review"] = df[text_col].apply(clean_for_sentiment)
df["review"] = df["review"].str.lower() # turn everything to lowercase
df[["review"]].head()
```

```python
df.to_csv(CLEANED_PATH)
```
